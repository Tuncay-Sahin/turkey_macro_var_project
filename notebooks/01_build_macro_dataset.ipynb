# Data Construction – Turkish Macroeconomic Dataset

This notebook builds the macroeconomic dataset used in the analysis.

Steps:
1. Download macroeconomic data from EVDS
2. Clean and standardize the series
3. Align monthly observations
4. Construct the final dataset

```python
import pandas as pd
import numpy as np

print("environment ready")
```

```python
from pathlib import Path

project_path = Path("..")

raw_path = project_path / "data" / "raw"
processed_path = project_path / "data" / "processed"

print(raw_path)
print(processed_path)
```

```python
# load processed datasets again

tge = pd.read_csv(processed_path / "tge_2019_2025.csv")
usd = pd.read_csv(processed_path / "usdtry_2019_2025.csv")
cpi = pd.read_csv(processed_path / "cpi_2019_2025.csv")
ipi = pd.read_csv(processed_path / "ipi_2019_2025.csv")
unemp = pd.read_csv(processed_path / "unemployment_2019_2025.csv")
bist = pd.read_csv(processed_path / "bist_2019_2025.csv")
```

```python
tge["date"] = pd.to_datetime(tge["date"])
usd["date"] = pd.to_datetime(usd["date"])
cpi["date"] = pd.to_datetime(cpi["date"])
ipi["date"] = pd.to_datetime(ipi["date"])
unemp["date"] = pd.to_datetime(unemp["date"])
bist["date"] = pd.to_datetime(bist["date"])
```

```python
macro_ext = tge

macro_ext = pd.merge(macro_ext, usd, on="date", how="inner")
macro_ext = pd.merge(macro_ext, cpi, on="date", how="inner")
macro_ext = pd.merge(macro_ext, ipi, on="date", how="inner")
macro_ext = pd.merge(macro_ext, unemp, on="date", how="inner")
macro_ext = pd.merge(macro_ext, bist, on="date", how="inner")
```
